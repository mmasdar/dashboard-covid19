```
import pandas as pd
import numpy as np
```

```
url_confirmed = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
url_deaths = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv'
url_recovered = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv'

df_confirmed = pd.read_csv(url_confirmed)
df_deaths = pd.read_csv(url_deaths)
df_recovered = pd.read_csv(url_recovered)
```

```
#Data total kejadian Covid-19 di Indonesia
indonesia_confirmed = df_confirmed.loc[df_confirmed['Country/Region'] == 'Indonesia']
indonesia_deaths = df_deaths.loc[df_deaths['Country/Region'] == 'Indonesia']
indonesia_recovered = df_recovered.loc[df_recovered['Country/Region'] == 'Indonesia']

#indonesia_confirmed = indonesia_confirmed.drop(['Province/State','Country/Region','Lat','Long'],axis=1).T
#indonesia_deaths = indonesia_deaths.drop(['Province/State','Country/Region','Lat','Long'],axis=1).T
#indonesia_recovered = indonesia_recovered.drop(['Province/State','Country/Region','Lat','Long'],axis=1).T

indonesia_confirmed = indonesia_confirmed.drop(['Province/State','Country/Region','Lat','Long'],axis=1).sum()
indonesia_deaths = indonesia_deaths.drop(['Province/State','Country/Region','Lat','Long'],axis=1).sum()
indonesia_recovered = indonesia_recovered.drop(['Province/State','Country/Region','Lat','Long'],axis=1).sum()

indonesia_confirmed
```

```
#GRAFIK PERKEMBANGAN COVID-19 HARIAN DI NEGARA TERTENTU
import matplotlib.pyplot as plt

x=indonesia_confirmed.index

fig = plt.figure(figsize= (80,20))
plt.plot(x, indonesia_confirmed, label='Positif', linewidth=5)
plt.plot(x, indonesia_deaths, label='Meninggal', linewidth=5)
plt.plot(x, indonesia_recovered, label='Sembuh', linewidth=5)

plt.legend(loc=2, fontsize=50)
```

```
#data penambahan kasus Covid-19 harian Indonesia
indonesia_confirmed_shift = (indonesia_confirmed - indonesia_confirmed.shift(1)).drop(indonesia_confirmed.index[0])
indonesia_deaths_shift = (indonesia_deaths - indonesia_deaths.shift(1)).drop(indonesia_deaths.index[0])
indonesia_recovered_shift = (indonesia_recovered - indonesia_recovered.shift(1)).drop(indonesia_recovered.index[0])

x_shift = indonesia_confirmed_shift.index
```

```
#GRAFIK PERKEMBANGAN COVID-19 HARIAN DI NEGARA TERTENTU
import matplotlib.pyplot as plt

fig = plt.figure(figsize= (80,20))
plt.plot(x_shift, indonesia_confirmed_shift, label='Positif', linewidth=5)
plt.plot(x_shift, indonesia_deaths_shift, label='Meninggal', linewidth=5)
plt.plot(x_shift, indonesia_recovered_shift, label='Sembuh', linewidth=5)

plt.legend(loc=2, fontsize=50)
```

```
#Menampilkan total kasus positif
indonesia_confirmed_total =  indonesia_confirmed[-1]

#Menampilkan total kasus sembuh
indonesia_recovered_total = indonesia_recovered[-1]

#Menampilkan total kasus kematian
indonesia_deaths_total = indonesia_deaths[-1]
```
